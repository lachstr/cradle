### Mathematical Framework


*''Newton's cradle can be modelled as mutilple pendulums confined to swing along a single axis which undergo elastic collisions with one another. The pendulums rest such that there is a small amount of horizontal displacement between them.''*

Let the cradle exist in a two dimensional cartesian coordinate system $(x,y)$

Then for each pendulum numbered $i \in \{1,2,3, ... , n\}$

- Has its own 1-dimensional co-ordinate system ($\theta_i$)
- Obeys differential equation $\ddot{\theta_i} = -\frac{g}{\ell}\sin(\theta_i)$
- Has length $\ell$, radius $r_i$, mass $m_i=\frac{4\pi\rho}{3}r_i^3$ where $\rho$ is a metal density.
- Has some $\Delta x_i$ along the $x$-axis, so they are 'almost touching' at rest.
 
The coordates of the $i$th pendulum in the cradel coordinate system can be found via the function;

$T(\theta_i, \Delta x_{i}) = (\ell\sin(\theta_i) + \Delta x_i, -\ell\cos(\theta_i))$

We will say a collision has occured if;

for any two $i, j$ $\in \{1,2,3, ... , n\}$

$$ |T(\theta_i, \Delta x_{i}) - T(\theta_j, \Delta x_{j})| \leq (r_i + r_j)$$ 

![](sketch2.jpg)

In the event of a collision it, will occur paralel to the $\hat{\theta}$ vector in each coordinate system - so we can use the 1 dimensional elastic collision formula. 

$$v_{1,f} = \frac{m_1-m_2}{m_1+m_2} v_{1,i} + \frac{2m_2}{m_1+m_2} v_{2, i}$$  
$$v_{2,f} = \frac{2m_1}{m_1+m_2} v_{1,i} + \frac{m_2-m_1}{m_1+m_2} v_{2, i}$$

Where $f,i$ refer to the final and initial velocities, $1,2$ refer to the 1st and 2nd bodies. 

In this context, the formula reduces to ($v=r\dot{\theta}$)
$$\dot{\theta}_{i,f} = \frac{m_i-m_j}{m_i+m_j} \dot{\theta}_{i,i} + \frac{2m_j}{m_i+m_j} \dot{\theta}_{j, i}$$  
$$\dot{\theta}_{j,f} = \frac{2m_i}{m_i+m_j} \dot{\theta}_{i,i} + \frac{m_j-m_i}{m_1+m_j} \dot{\theta}_{j, i}$$

(NB: The $\dot{\theta}_{i,i}$  refers to the initial $\dot{\theta}$ value of the $i$th pendulum)
